### 3. Audience Study:
- Analyze the distribution of streamers' audiences by country. Are there regional preferences for specific content categories?

```python
%%sql

SELECT Categories,
    Country,
    SUM(Suscribers) AS Total_Subscribers
FROM youtubers_df
GROUP BY Categories, Country
ORDER BY Categories, SUM(Suscribers) DESC;
```

There is a prefence for children related content in the United States, categories such as AnimaciÃ³n, Videojuegos, Humor, Juguetes (animation, videogames, humor, toys) have a high number of subscribers from the United States.
There is also a preference for Deportes (sports) in the United States and India has a high number of subscribers for PelÃ­culas, MÃºsica y baile (movies, music and dance) content.

### 4. Performance Metrics:
- Calculate and visualize the average number of subscribers, visits, likes, and comments.
- Are there patterns or anomalies in these metrics?

```python
%%sql

SELECT
    DISTINCT ROUND(AVG(Suscribers)) AS Avg_Subscibers,
    ROUND(AVG(Visits)) AS Avg_Visits,
    ROUND(AVG(Likes)) AS Avg_Likes,
    ROUND(AVG(Comments)) AS Avg_Comments
    FROM youtubers_df
    WHERE Visits != 0;
```

The average number of subscribers is more than the average number of visits, this could suggest that most of the content are not very engaging to the subscribers.

The line chart below also shows a positive correlation between the number of likes and comments



### 5. Content Categories:
- Explore the distribution of content categories. Which categories have the highest number of streamers?

```python
%%sql

SELECT
    Categories,
    COUNT(Categories) AS Number_of_streamers
FROM youtubers_df
GROUP BY Categories
ORDER BY COUNT(Categories) DESC
LIMIT 10;
```

```python
%%sql

SELECT Categories,
    ROUND(AVG(Suscribers)) AS Avg_Subscibers,
    ROUND(AVG(Visits)) AS Avg_Visits,
    ROUND(AVG(Likes)) AS Avg_Likes,
    ROUND(AVG(Comments)) AS Avg_Comments
    FROM youtubers_df
    GROUP BY Categories
    ORDER BY AVG(Suscribers) DESC
    LIMIT 20;
```

### 6. Benchmarking:
- Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments.- Who are the top performing content creators?

Top 10 performing content creators with above-average performance in terms of subscribers, visits, likes, and comments; ordered by the highest number of subscribers

```python
%%sql

SELECT *
FROM youtubers_df
WHERE Suscribers > (SELECT AVG(Suscribers)
                    FROM youtubers_df)
AND Visits > (SELECT AVG(Visits)
              FROM youtubers_df
             WHERE Visits != 0)
AND Likes > (SELECT AVG(Likes)
             FROM youtubers_df
             WHERE Visits != 0)
AND Comments > (SELECT AVG(Comments)
                FROM youtubers_df
                WHERE Visits != 0)
LIMIT 10
;
```
